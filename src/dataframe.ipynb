### Text Pre-processing

```python
import pandas as pd
import sys
import csv
import re
maxInt = sys.maxsize

while True:
    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
df_extracted = pd.read_csv('df_extracted.csv', engine='python', encoding='utf-8', error_bad_lines=False)
```

```python
df_extracted
```

```python
#Remove .txt extension on the file_name column
df_extracted['file_name'] = df_extracted['file_name'].str.replace(r'.txt$', '')
df_extracted
```

```python
# Lower all text contents
df_extracted['text'] = df_extracted['text'].str.lower()
df_extracted
```

```python
# replace special special character
df_extracted['text'] = df_extracted['text'].str.replace("\n","").replace("\t"," ").replace("\x0c"," ").replace(".",". ").replace(',',"").replace("\r\n"," ")
df_extracted['text'] = df_extracted['text'].str.replace("\r"," ")
df_extracted
```

```python
df_extracted.to_csv('df_extracted_cleaned.csv')
```
