### Text Pre-processing

```python
import pandas as pd
import sys
import csv
import re
maxInt = sys.maxsize

while True:
    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
df_extracted = pd.read_csv('df_extracted.csv', engine='python', encoding='utf-8', error_bad_lines=False)
```

```python
df_extracted.describe()
```

```python
df_extracted
```

```python
#Remove .txt extension on the file_name column
df_extracted['file_name'] = df_extracted['file_name'].str.replace(r'.txt$', '')
df_extracted
```

```python
# Lower all text contents
df_extracted['text'] = df_extracted['text'].str.lower()
df_extracted
```

```python
#drop empty row of the text column
df_extracted.dropna(subset = ["text"], inplace=True)
```

```python
df_extracted
```

```python
# replace special special character
df_extracted['text'] = df_extracted['text'].str.replace("\n"," ").replace("\t"," ").replace("\x0c"," ").replace(".",". ").replace(',',"").replace("\r\n"," ")
df_extracted['text'] = df_extracted['text'].str.replace("\r"," ")
df_extracted
```

```python
df_extracted_1 = df_extracted[:5000]
df_extracted_2 = df_extracted[5000:10000]
df_extracted_3 = df_extracted[10000:15000]
df_extracted_4 = df_extracted[15000:]
```

```python
df_extracted_1.to_csv('df_extracted_1.csv')
df_extracted_2.to_csv('df_extracted_2.csv')
df_extracted_3.to_csv('df_extracted_3.csv')
df_extracted_4.to_csv('df_extracted_4.csv')
```
